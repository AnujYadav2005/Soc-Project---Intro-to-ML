## Linear Regression

```python
# define an array x of atleast 100 points with a separation of at least 1

x = np.linspace(0, 100, 100)  # 100 points from 0 to 100 (inclusive)

# now choose any 2 values for A, B and generate an array y_true such that y_true = Ax + B for all x defined above

A = 2.0  # slope
B = 3.0  # intercept

y_true = A * x + B
```

```python
# Add Gaussian Noise which can vary from upto -20 to 20 to y_true to get our data

noise = np.random.normal(0, 20, size=len(x))
y_noisy = y_true + noise
```

```python
# print the data, y_true and x

print("x values:")
print(x)
print("True y values:")
print(y_true)
print("Noisy y values:")
print(y_noisy)
```

```python
# plot scatterplot between data and x
# also plot y_true vs x on the same plot
plt.figure(figsize=(8, 6))
plt.scatter(x, y_noisy, label='Noisy data points')
plt.plot(x, y_true, color='r', label='True line: y = {}x + {}'.format(A, B))
plt.title('Linear Regression Example with Gaussian Noise')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()



```

```python
# write a function definition to calculate loss of linear Regression

def calculate_loss(y_true, y_pred):
    return np.mean((y_true - y_pred)**2)



```

```python
# write a function definition to calculate derivative of cost function of linear regression with respect to parameter A


def derivative_wrt_A(x, y_true, y_pred):
    return np.mean((y_pred - y_true) * x)



```

```python
# write a function definition to calculate derivative of cost function of linear regression with respect to parameter B

def derivative_wrt_B(y_true, y_pred):
    return np.mean(y_pred - y_true)



```

```python
# define A = 0 and B = 0 randomly
# also define 2 lists to store values of A and B at various iterations

np.random.seed(0)
A = np.random.randn()
B = np.random.randn()

A_values = []
B_values = []



```

```python
# compute the model array based on initial values of A and B
# Also print the cost function for the data before training

y_pred = A * x + B
initial_cost = np.mean((y_pred - y_noisy)**2)
print("\nInitial Cost (Mean Squared Error) before training:", initial_cost)

```

```python
# Define the Learning rate, alpha randomly. You will have to modify it during training
alpha = 0.0001
```

```python
# write the main training loop and train for atleast 2000 iterations
# print the cost at each iteration while training
# save the values of A and B at atleast 5 different iterations spaced by atleast 50 into the lists defined earlier
# Also, make sure you use the same model to find derivative of cost wrto A and B and then modify A and B together

# Training loop
iterations = 2000
print("\nTraining...")
for i in range(iterations):
    # Compute current predictions
    y_pred = A * x + B

    # Calculate gradients
    dA = np.mean((y_pred - y_noisy) * x)
    dB = np.mean(y_pred - y_noisy)

    # Update parameters A and B
    A -= alpha * dA
    B -= alpha * dB

    # Store A and B values at certain iterations
    if i % 50 == 0:
        A_values.append(A)
        B_values.append(B)

    # Print cost at every 100 iterations
    if i % 100 == 0:
        current_cost = np.mean((y_pred - y_noisy)**2)
        print("Iteration {}: Cost = {}".format(i, current_cost))

print("\nTraining completed.")







```

## Logistic Regression

Using the following data, let's train a Logistic Regression model. I have already plotted the data for visualization, noting that I intentionally included some overlapping examples to reflect real-world scenarios where a clear distinguishing line isn't always present.

```python
x2_1 = np.random.randint(30, 70, 50)
x1_1 = np.random.randint(15, 50, 50)
x1_0 = np.random.randint(45, 90, 50)
x2_0 = np.random.randint(5, 45, 50)

# we have to make a single vector for each of x1 and x2 by combining both the classes for training
x1 = np.hstack((x1_0, x1_1))
x2 = np.hstack((x2_0, x2_1))
# dividing both x1 and x2 by large number like 1000 to make sure that their sigmoid values don't saturate to either 0 or 1
x1 = x1/1000
x2 = x2/1000

# creating data
y0 = np.zeros(50)
y1 = np.ones(50)
y = np.hstack((y0, y1))

plt.scatter(x1_1, x2_1, color="blue")
plt.scatter(x1_0, x2_0, color="red")
plt.show()
```

## Line we are considering as a distinguishing line is-

$$
_0 +_1*x_1 +_2*x_2 = 0
$$

```python
# sigmoid function

def sigmoid(z):
    return 1 / (1 + np.exp(-z))


```

```python
# function to compute cost of Logistic Regression


def compute_cost(theta0, theta1, theta2, x1, x2, y):
    m = len(y)
    z = theta0 + theta1 * x1 + theta2 * x2
    h = sigmoid(z)
    cost = (-1 / m) * np.sum(y * np.log(h) + (1 - y) * np.log(1 - h))
    return cost

```

```python
# function to compute the derivative of cost with respect to theta1


def derivative_wrt_theta1(theta0, theta1, theta2, x1, x2, y):
    m = len(y)
    z = theta0 + theta1 * x1 + theta2 * x2
    h = sigmoid(z)
    d_theta1 = (1 / m) * np.sum((h - y) * x1)
    return d_theta1

```

```python
# function to compute the derivative of cost with respect to theta2

def derivative_wrt_theta2(theta0, theta1, theta2, x1, x2, y):
    m = len(y)
    z = theta0 + theta1 * x1 + theta2 * x2
    h = sigmoid(z)
    d_theta2 = (1 / m) * np.sum((h - y) * x2)
    return d_theta2


```

```python
# function to compute the derivative of cost with respect to theta0


def derivative_wrt_theta0(theta0, theta1, theta2, x1, x2, y):
    m = len(y)
    z = theta0 + theta1 * x1 + theta2 * x2
    h = sigmoid(z)
    d_theta0 = (1 / m) * np.sum(h - y)
    return d_theta0


```

```python
# randomly initialize theta0, theta1, theta2 and compute and print the initial cost



np.random.seed(0)
theta0 = np.random.randn()
theta1 = np.random.randn()
theta2 = np.random.randn()

initial_cost = compute_cost(theta0, theta1, theta2, x1, x2, y)
print("Initial Cost (Before training):", initial_cost)
```

```python
# learning rate, alpha. You will have to modify it while training
alpha = 0.01
```

```python
# main training loop
# Make sure you use the same model to find derivative of cost wrto theta0, theta1 and theta2 and then modify them together

iterations = 1000
for i in range(iterations):
    # Compute gradients
    d_theta1 = derivative_wrt_theta1(theta0, theta1, theta2, x1, x2, y)
    d_theta2 = derivative_wrt_theta2(theta0, theta1, theta2, x1, x2, y)
    d_theta0 = derivative_wrt_theta0(theta0, theta1, theta2, x1, x2, y)

    # Update parameters using gradient descent
    theta1 -= alpha * d_theta1
    theta2 -= alpha * d_theta2
    theta0 -= alpha * d_theta0

    # Compute and print the cost every 100 iterations
    if i % 100 == 0:
        current_cost = compute_cost(theta0, theta1, theta2, x1, x2, y)
        print(f"Iteration {i}: Cost = {current_cost}")

print("\nTraining completed.")







```

Modify the values of theta_1 and theta0 for ploting;

Read below:

Earlier, we scaled down both `x_1` and `x_2` by 1000 for easier training. Now, we need to multiply both sides by 1000.

So,

$$
1000x_2 = \left(\frac{-1000 \theta_1}{\theta_2}\right) x_1 + \left(\frac{-1000 \theta_0}{\theta_2}\right)
$$

This simplifies to:

$$
y = \theta_{1,\text{new}} x_1 + \theta_{0,\text{new}}
$$

Finally, plot `y` vs `1000x_1` using the plot function.


```python

# Calculate the decision boundary
theta0_original = theta0
theta1_original = theta1 * 1000
theta2_original = theta2 * 1000

x1_decision_boundary = np.linspace(0, 100, 100)
x2_decision_boundary = (-theta0_original - theta1_original * x1_decision_boundary) / theta2_original





```

```python
# plot the final line obtained after training along with the initial data on the same plot

# Plotting decision boundary and data points
plt.figure(figsize=(8, 6))
plt.scatter(x1_1, x2_1, color="blue", label='Class 1')
plt.scatter(x1_0, x2_0, color="red", label='Class 0')
plt.plot(x1_decision_boundary, x2_decision_boundary, color='green', label='Decision Boundary')
plt.title('Logistic Regression Decision Boundary')
plt.xlabel('x1')
plt.ylabel('x2')
plt.legend()
plt.grid(True)
plt.show()

# Print final parameters
print("\nFinal parameters:")
print(f"theta0 = {theta0_original}")
print(f"theta1 = {theta1_original}")
print(f"theta2 = {theta2_original}")






```
